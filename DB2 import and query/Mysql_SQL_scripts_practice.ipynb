# Problem 2: Find average college enrollments by community area

```python
sql4 = 'select Community_Area_Name,avg(College_Enrollment_number_of_students_) as CEN_avg \
from CHICAGO_PUBLIC_SCHOOLS group by Community_Area_Name;'
df = pd.read_sql(sql4, con=db)
df
```

# Problem 3: Find the number of schools that are healthy school certified

```python
sql5 = "select count(Healthy_Schools_Certified) as health_school_count from CHICAGO_PUBLIC_SCHOOLS \
where Healthy_Schools_Certified = 'YES';"
df = pd.read_sql(sql5, con=db)
df
```

# Problem 4: How many observations have a LocationDescription value of GAS STATION?

```python
sql6 = "select count(*) as Gas_Station_Counts from CHICAGO_CRIME_DATA where Location_Description = 'GAS STATION';"
df = pd.read_sql(sql6, con=db)
df
```

# Problem 5: Retrieve a list of the top 10 community areas which have most number of schools and sorted in descending order.

```python
sql7 = "select Community_Area_Name,count(Name_of_School) as school_count \
from CHICAGO_PUBLIC_SCHOOLS \
group by Community_Area_Name \
order by school_count DESC limit 10"
df = pd.read_sql(sql7, con=db)
df
```

# Problem 6: How many observations have value MOTOR VEHICLE THEFT in the Primary Type variable (this is the number of crimes related to Motor vehicles)

```python
sql8 = 'select count(*) as observations from CHICAGO_CRIME_DATA \
where Primary_Type = "MOTOR VEHICLE THEFT";'
df = pd.read_sql(sql8, con=db)
df
```

# Problem 7: Find the minimum “Average Student Attendance” for community are where hardship is 96

```python
sql9 = "select MIN(b.Average_Student_Attendance) from CENSUS_DATA a,CHICAGO_PUBLIC_SCHOOLS b \
where a.HARDSHIP_INDEX = 96 \
and a.Community_Area_Number = b.Community_Area_Number;"
df = pd.read_sql(sql9, con=db)
df
```
