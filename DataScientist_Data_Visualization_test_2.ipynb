```python
url = 'Topic_Survey_Assignment.csv'
import pandas as pd
dfs = pd.read_csv(url,index_col=0)
dfs
```

```python
dfs.sort_values('Very interested',ascending=False,inplace=True)
dfs
```

```python
dfs['total']=dfs.sum(axis=1)
dfs['Very interested']=dfs['Very interested']/dfs['total']
dfs['Somewhat interested']=dfs['Somewhat interested']/dfs['total']
dfs['Not interested']=dfs['Not interested']/dfs['total']
dfs
```

```python
dfs.drop('total',axis=1,inplace=True)
```

```python
dfs
```

```python
dfs2 = dfs.round(2)
dfs2.index
dfs2['Very interested']
```

```python

# -*- coding: utf-8 -*-
import matplotlib.pyplot as plt

name_list = dfs2.index
num_list1 = dfs2['Very interested']
num_list2 = dfs2['Somewhat interested']
num_list3 = dfs2['Not interested']
plt.figure(figsize=(20,8))
x =list(range(len(num_list1)))
total_width, n = 0.8, 3
width = total_width / n
 
plt.bar(x, num_list1, width=width, label='Very interested', fc = '#5cb85c')
for i in range(len(x)):
    x[i] = x[i] + width
plt.bar(x, num_list2, width=width, label='Somewhat interested', fc = '#5bc0de')
plt.xticks(x, name_list,fontsize=14)
for i in range(len(x)):
    x[i] = x[i] + width
plt.bar(x, num_list3, width=width,  label='Not interested', fc = '#d9534f')
plt.title("ABC",fontdict={'fontsize': 16})
plt.legend(fontsize=14)
plt.show()
```
